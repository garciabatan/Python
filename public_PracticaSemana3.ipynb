```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('iris.csv')
```

```python
df
```

```python
print(df['Species'].value_counts())
```

```python
df.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
```

```python
df.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].values.reshape(-1,1).mean()
```

```python
print('Promedio Largo Petalo:',df['Petal.Length'].mean())
print('Promedio Ancho Petalo:',df['Petal.Width'].mean())
print('Promedio Largo Sepalo:',df['Sepal.Length'].mean())
print('Promedio Ancho Sepalo:',df['Sepal.Width'].mean())
```

```python
max = df.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].values.reshape(-1,1).max() 
min = df.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].values.reshape(-1,1).min()

rango = max - min
print('Rango de los datos:',rango)
```

```python
print('Rango Largo Petalo:',df['Petal.Length'].max() - df['Petal.Length'].min())
print('Rango Ancho Petalo:',df['Petal.Width'].max() - df['Petal.Width'].min())
print('Rango Largo Sepalo:',df['Sepal.Length'].max()- df['Sepal.Length'].min())
print('Rango Ancho Sepalo:',df['Sepal.Width'].max() - df['Sepal.Length'].min())
```

```python
df.loc[:,['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']].values.reshape(-1,1).std()
```

```python
print('Promedio Largo Petalo:',df['Petal.Length'].std())
print('Promedio Ancho Petalo:',df['Petal.Width'].std())
print('Promedio Largo Sepalo:',df['Sepal.Length'].std())
print('Promedio Ancho Sepalo:',df['Sepal.Width'].std())
```

## Analisis Mediante Visualizaciones

```python
df.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(8,4)
plt.show()
```

### Analisis dispersion de cada variable

```python
sns.catplot(x="Species", y="Petal.Length", hue="Species", data=df)
plt.show()
```

```python
sns.catplot(x="Species", y="Petal.Width", hue="Species", data=df)
plt.show()
```

```python
sns.catplot(x="Species", y="Sepal.Length", hue="Species", data=df)
plt.show()
```

```python
sns.catplot(x="Species", y="Sepal.Width", hue="Species", data=df)
plt.show()
```

### Analisis Mediante Boxplot

```python
sns.catplot(x="Species", y="Petal.Length", hue="Species", data=df, kind='box')
plt.show()
```

```python
sns.catplot(x="Species", y="Petal.Width", hue="Species", data=df, kind='box')
plt.show()
```

### Analisis por medio de violines

```python
sns.violinplot(x="Species", y="Petal.Length", hue="Species", data=df)
plt.show()
```

```python
sns.violinplot(x="Species", y="Petal.Width", hue="Species", data=df)
plt.show()
```

```python
sns.jointplot(x='Sepal.Length',y='Sepal.Width',data=df,size=5)
```
